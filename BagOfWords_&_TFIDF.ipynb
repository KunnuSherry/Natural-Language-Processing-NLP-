```python
paragraph = """Salman Khan is an Indian actor and producer, known for his work in Hindi films. He made his film debut with a brief role in Biwi Ho To Aisi (1988), before having his breakthrough with Sooraj Barjatya's blockbuster romance Maine Pyar Kiya (1989) that won him the Filmfare Award for Best Male Debut.[1] In the early 1990s, he earned success with the action film Baaghi: A Rebel for Love (1990) and the romance Sanam Bewafa, Saajan (both 1991).[2][3][4]

The success of the family drama Hum Aapke Hain Koun..! (1994) and the melodramatic action Karan Arjun (1995) revitalised Khan's career and established him in Bollywood.[5][6][7] Also in 1994, he co-starred with Aamir Khan in the comedy movie Andaz Apna Apna, which was poorly received at that time, but later became a cult film in India.[8] Among his three film releases of 1996 were Sanjay Leela Bhansali's critically acclaimed musical drama Khamoshi and the Raj Kanwar-directed drama Jeet.[9] The following year, he played dual roles in David Dhawan's comedy Judwaa. In 1998, Khan featured in Sohail Khan's Pyaar Kiya To Darna Kya, and appeared briefly in the romantic drama Kuch Kuch Hota Hai, both of which ranked among the top-earning Bollywood productions of 1998.[10] For the latter, he was awarded the Filmfare Award for Best Supporting Actor.[11]"""
```

```python
paragraph # this is the corpus
```

```python
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
#Tokenization - converts paras to sentences to words
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
sentences=nltk.sent_tokenize(paragraph) # converts paras to sentence
```

```python
sentences
```

```python
type(sentences)
```

```python
stemmer=PorterStemmer()
```

```python
stemmer.stem('history')
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
import nltk
nltk.download('wordnet')
lemmatizer.lemmatize('goes')
```

```python
## Remove special characters

import re
corpus=[]
for i in range(len(sentences)):
  review=re.sub('[^a-zA-Z]',' ',sentences[i])
  review=review.lower()
  corpus.append(review)
```

```python
corpus
```

```python
## stemming
import nltk
nltk.download('stopwords')
for i in corpus:
  words=nltk.word_tokenize(i)
  for word in words:
    if word not in set(stopwords.words('english')):
      print(stemmer.stem(word))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(binary=True, ngram_range=(3,3))
```

```python
X= cv.fit_transform(corpus)
```

```python
cv.vocabulary_ #these are the list of words index - feature number
```

```python
#TFIDF
from sklearn.feature_extraction.text import TfidfVectorizer
cv=TfidfVectorizer(ngram_range=(3,3)) #max_features=3 given top3 max features
X=cv.fit_transform(corpus)
```
